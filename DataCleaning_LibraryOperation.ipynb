```python
import pandas as pd
import numpy as np
```

```python
UL1 = pd.read_csv("LOW February 2017 Suma Initiative 1 - Sheet1.csv",delimiter = ',')
UH2 = pd.read_csv("HIGH April 2017 Suma Initiative 2 - All Data.csv",delimiter = ',')
UL3 = pd.read_csv("LOW July 2017 Suma Initiative 3.csv",delimiter = ',')
UM4 = pd.read_csv("MEDIUM January 2018 Suma Initiative 5.csv - January 2018 Suma Data.csv.csv",delimiter = ',')
UM5 = pd.read_csv("MEDIUM November 2017 Suma Initiative 4.csv",delimiter = ',')
UM6 = pd.read_csv("HIGH April 2018 Suma Initiative 6.csv",delimiter = ',')
UM7 = pd.read_csv("MEDIUM Suma Initative Week November 2018.csv",delimiter = ',')
```

```python
##### Data Cleaning Function
# Imput UL1...UM7 to the following function

#H2 data
def CleanH2(data):
    af = data.drop(columns=[ 'Count Number','Time'])
    df = af['Date'].apply(lambda s:s.replace('"', ""))
    df['Date'] = pd.to_datetime(af['Date'])
    df['day_of_week'] = df['Date'].dt.weekday_name
    return df.to_csv('H2.csv', index = False)

# L1data
def CleanL1(data):
    af = data.drop(columns=[ 'Count Number','Minute', 'Second'])
    af['Date'] = pd.to_datetime(af['Date'])
    af['day_of_week'] = af['Date'].dt.weekday_name
    return af.to_csv('L1.csv', index = False)

```

```python
def namestr(obj, namespace):
    return [name for name in namespace if namespace[name] is obj]

##
def CleanM45H6(data):
    af = data.drop(columns=[ 'Session Start', 'Count ID','Session ID'])
    # rename the columb
    af.rename(columns={
                 'I see...: Individual(s)': 'Individual',
                 'I see...: a_Group': 'Group',
                 'they are...: Talking': 'Talking',
                 'they are...: Sleeping': 'Sleeping',
                 'sitting at/on...: a Large Table (seats 4+)': 'Large Table (seats 4+)',
                 'sitting at/on...: Hard Seating': 'Hard Seating',
                'sitting at/on...: a Small Table (seats <4)': 'Small Table (seats <4)',
    'sitting at/on...: Soft Seating': 'Soft Seating',
                 'sitting at/on...: Rolling Table': 'Rolling Table',
                 'sitting at/on...: Rolling Seating': 'Rolling Seating',
                 'with...: Personal tech (laptop, phone, etc.)': 'Personal tech',
                 'with...: Books, papers, notebooks, etc.': 'Books, papers, notebooks',
                 'with...: Library tech (computer station, vizhub, etc.)': 'Library tech',
                'with...: Whiteboard': 'Whiteboard',
                'they are...: Focused (looking at screen, book, etc.)': 'Focused',
                'No Activity: No Activity': 'No Activity',
                'I see...: Small Group (3-4)': 'Small Group (3-4)',
                'I see...: Large Group (5+)': 'Large Group (5+)',
                'Noise Level: Silent': 'Silent',
                'Noise Level: Quiet': 'Quiet',
                'Noise Level: Animated': 'Animated',
                'Noise Level: Conversational': 'Conversational',
                     'with...: Digital Screen': 'Digital Screen',
                'with...: Computer Station': 'Computer Station',
                'I see...: New Activity': 'New Activity',
                'they are...: Walking': 'walking',
                'they are...: Talking': 'Talking',
                'sitting at/on...: a_Carrel': 'Sit Carrel',
                'sitting at/on...: Sitting on the ground': 'Sit ground'}, inplace = True)
    # Split
    splitlist = af['Count Date'].str.split(' ', 1, expand=True)
    splitlist.columns = ['date', 'time']
    timesplit = splitlist['time'].str.split(':', 1, expand=True)
    timesplit.columns = ['hour', 'minute']
    Locationsplit = af['Location'].str.split('-', 1, expand=True)
    Locationsplit.columns = ['building', 'min']
    Locationsplit1 = Locationsplit['min'].str.split('(', 1, expand=True)
    Locationsplit2 = Locationsplit1[1].str.split(')', 1, expand=True)
    Locationsplit3 = Locationsplit2[0].str.split(':', 1, expand=True)
    # match the cleaned cell to var
    af['Date'] = splitlist['date']
    af['hour'] = timesplit['hour']
    af['Sublocation'] = Locationsplit1[0]
    af['Building'] = Locationsplit['building']
    af['Location ID'] = Locationsplit3[1]
    af['Date'] = pd.to_datetime(af['Date'])
    af['day_of_week'] = af['Date'].dt.weekday_name
    af1 = af.drop(columns=[ 'Count Date'])
    
    return af1.to_csv(namestr(data, globals())[0] +'.csv', index = False)

#
```
